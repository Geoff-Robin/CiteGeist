```python
import pandas as pd
df = pd.read_csv('dataset/cs_papers_api.csv')
df.describe()
df.head()
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
from datasketch import MinHash, MinHashLSH
# Combine title and abstract for LSH
df['combined_text'] = df['title'] + ' ' + df['abstract']

# Initialize TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['combined_text'])

# Create MinHash for each article
minhashes = []
for i in range(tfidf_matrix.shape[0]):
    minhash = MinHash()
    for j in tfidf_matrix[i].nonzero()[1]:  # Get indices of non-zero elements
        minhash.update(str(j).encode('utf8'))  # Use indices as features for MinHash
    minhashes.append(minhash)

# Build LSH
lsh = MinHashLSH(threshold=0.5, num_perm=128)  # Threshold can be adjusted based on your needs
for i, minhash in enumerate(minhashes):
    lsh.insert(f'article_{i}', minhash)

# Querying similar articles
query_index = 0  # Index of the article you want to find similar articles for
query_minhash = minhashes[query_index]
similar_indices = lsh.query(query_minhash)

# Display results
print("Similar Articles to:", df['title'][query_index])
for index in similar_indices:
    print(" -", df['title'][int(index.split('_')[1])])
```
