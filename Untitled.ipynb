```python
import json
import os
import time
import csv
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import random
from collections import defaultdict

def load_or_create_sample(file_path, sample_size=5000, sample_file='sampled_cs_papers_5000.csv'):
    if os.path.exists(sample_file):
        print(f"Loading sampled papers from {sample_file}")
        with open(sample_file, 'r', newline='', encoding='utf-8') as f:
            reader = csv.DictReader(f)
            return list(reader)

    print(f"Sampling {sample_size} papers from the dataset...")
    cs_papers = []
    category_counts = defaultdict(int)
    total_cs_papers = 0

    # First pass: count papers in each CS category
    with open(file_path, 'r') as file:
        for line in file:
            paper = json.loads(line)
            if 'cs.' in paper['categories']:
                total_cs_papers += 1
                primary_category = paper['categories'].split()[0]
                category_counts[primary_category] += 1

    # Calculate sampling ratios
    sampling_ratios = {cat: count / total_cs_papers for cat, count in category_counts.items()}

    # Second pass: sample papers maintaining category ratios
    sampled_papers = defaultdict(list)
    with open(file_path, 'r') as file:
        for line in file:
            paper = json.loads(line)
            if 'cs.' in paper['categories']:
                primary_category = paper['categories'].split()[0]
                if len(sampled_papers[primary_category]) < sampling_ratios[primary_category] * sample_size:
                    sampled_papers[primary_category].append(paper)

    # Combine all sampled papers
    for papers in sampled_papers.values():
        cs_papers.extend(papers)

    # Shuffle to avoid any ordering bias
    random.shuffle(cs_papers)
    cs_papers = cs_papers[:sample_size]

    # Save the sampled papers to CSV
    fieldnames = ['id', 'title', 'categories', 'abstract']
    with open(sample_file, 'w', newline='', encoding='utf-8') as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        for paper in cs_papers:
            writer.writerow({
                'id': paper['id'],
                'title': paper['title'],
                'categories': paper['categories'],
                'abstract': paper['abstract']
            })

    return cs_papers

def preprocess_data(data):
    return [f"{paper['title']} {paper['abstract']}" for paper in data]

def create_feature_vectors(preprocessed_data):
    vectorizer = TfidfVectorizer(stop_words='english')
    return vectorizer.fit_transform(preprocessed_data), vectorizer

def calculate_similarity(feature_vectors, query_vector):
    return cosine_similarity(query_vector, feature_vectors).flatten()

def get_recommendations_based_on_keyword(keyword, vectorizer, feature_vectors, data, top_n=5, score_threshold=0.5):
    # Transform the keyword into a vector
    query_vector = vectorizer.transform([keyword])

    # Calculate similarity
    similarity_scores = calculate_similarity(feature_vectors, query_vector)
    
    # Filter papers with scores above the threshold
    filtered_indices = [i for i, score in enumerate(similarity_scores) if score >= score_threshold]
    
    # Sort the filtered papers by score
    filtered_indices.sort(key=lambda i: similarity_scores[i], reverse=True)
    
    # Limit to the top N results
    top_indices = filtered_indices[:top_n]
    
    return [(data[i]['id'], data[i]['title'], data[i]['categories'], similarity_scores[i]) for i in top_indices]

def main():
    start_time = time.time()

    # Set the file paths
    file_path = os.path.join(os.path.expanduser('~'), 'Downloads', 'arxiv-metadata-oai-snapshot.json')
    sample_file = 'sampled_cs_papers_5000.csv'
    
    # Load or create sampled data
    data = load_or_create_sample(file_path, sample_size=5000, sample_file=sample_file)
    print(f"Number of sampled CS papers: {len(data)}")

    # Preprocess data
    print("Preprocessing data...")
    preprocessed_data = preprocess_data(data)

    # Create feature vectors and get the vectorizer
    print("Creating feature vectors...")
    feature_vectors, vectorizer = create_feature_vectors(preprocessed_data)
    
    # Get a keyword from the user
    keyword = input("Enter a keyword to search for similar research papers: ").strip()
    
    # Set the similarity score threshold
    score_threshold = float(input("Enter the minimum relevance score (e.g., 0.5): ").strip())

    # Get recommendations based on the keyword
    recommendations = get_recommendations_based_on_keyword(keyword, vectorizer, feature_vectors, data, score_threshold=score_threshold)
    
    if recommendations:
        print(f"\nRecommendations based on the keyword: '{keyword}' with relevance scores above {score_threshold}:")
        for id, title, categories, score in recommendations:
            print(f"ID: {id}, Title: {title}")
            print(f"Categories: {categories}, Relevance Score: {score:.2f}")
            print()
    else:
        print(f"\nNo papers found with a relevance score above {score_threshold} for the keyword '{keyword}'.")

    end_time = time.time()
    print(f"\nTotal execution time: {end_time - start_time:.2f} seconds")

if __name__ == "__main__":
    main()
```
